# 2. 데이터 전처리

## 2.1 모듈 및 학습 데이터 로딩

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

cust_df = pd.read_csv('./data/train_santander.csv', encoding='latin-1')
print('dataset shape : ', cust_df.shape)
cust_df.head(3)
```

## 2.3 TARGET 피처의 데이터 분포 확인

## 2.5 데이터 세트 분리

클래스 데이터 세트와 피처 데이터 세트를 분리

```python
# 피처 세트와 레이블 세트 분리. 레이블 컬럼은 DataFrame의 맨 마지막에 위치해 컬럼 위치 -로 분리
X_features = cust_df.iloc[:,:-1]
y_labels = cust_df.iloc[:,-1]
print('피처 데이터 shape : {0}'.format(X_features.shape))
```

<br>

학습과 성능 평가를 위해 원본 데이터 세트를 학습 데이터 세트와 테스트 데이터 세트로 분리  
비대칭한 데이터 세트이므로 클래스인 TARGET 값 분포도가 학습 데이터와 테스트 데이터 세트에 모두 비슷하게 추출됐는 지 확인

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels,
                                                    test_size=0.2, random_state=0)
print('학습 세트 Shape : {0}, 테스트 세트 Shape : {1}'.format(X_train.shape, X_test.shape))

train_cnt = y_train.count()
test_cnt = y_test.count()
print('학습 세트 레이블 값 분포 비율')
print(y_train.value_counts()/train_cnt)
print('테스트 세트 레이블 값 분포 비율')
print(y_test.value_counts()/test_cnt)
```

<br>

# 3. XGBoost 모델 학습과 하이퍼 파라미터 튜닝

## 3.1 모델 학습

사이킷런 래퍼 XGBClassifier 사용  
`n_estimators = 500`  
`early_stopping_rounds = 100`  
`eval_metric = 'auc'`  
`eval_set=[(X_train, y_train), (X_test, y_test)]`  
(사실 테스트 데이터 세트를 XGBoost의 평가 데이터 세트로 사용하면 과적합 가능성을 증가시킬 수 있다.)

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# n_estimators는 500으로, random_state는 예제 수행 시마다 동일한 예측 결과를 위해 설정
xgb_clf = XGBClassifier(n_estimators=500, random_state=156)

# 성능 평가 지표를 auc로, 조기 중단 파라미터는 100으로 설정하고 학습 수행
xgb_clf.fit(X_train, y_train,
            early_stopping_rounds=100,
            eval_metric='auc',
            eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:, 1], average='macro')
print('ROC AUC : {0:.4f}'.format(xgb_roc_score))
```

$\rightarrow$ 테스트 데이터 세트로 예측 시 ROC AUC는 약 0.8419

<br>

## 3.2 하이퍼 파라미터 튜닝

컬럼의 개수가 많으므로 과적합 가능성을 가정  
`max_depth`, `min_child_weight`, `colsample_bytree` 하이퍼 파라미터만 일차 튜닝 대상으로 함

<br>

**하이퍼 파라미터 튜닝 요령**

- 먼저 2~3개 정도의 파라미터를 결합해 최적 파라미터를 찾아냄
- 그런 다음 이 최적 파라미터를 기반으로 다시 1~2개 파라미터를 결합해 파라미터 튜닝을 수행

<br>

다음 코드는 8개의 하이퍼 파라미터경우의 수를 가짐 ($2 \times 2 \times 2 = 8$)  
  
  
수행 시간이 오래 걸리므로 `n_estimators`는 100으로 줄이고, `early_stopping_rounds`도 30으로 줄여 테스트  
나중에 하이퍼 파라미터 튜닝이 완료되면 다시 증가시켜서 모델 학습 수행  
  
  
또한 수행 시간을 줄이기 위해 `GridSearchCV`의 `cv` 파라미터도 설정하지 않음  
(`cv` 파라미터가 설정되지 않으면 하이퍼 파라미터만 번갈아가면서 평가를 수행  
  
  
`GridSearchCV` 적용 후에 최적 파라미터가 어떤 것이고, `GridSearchCV`에서 재학습된 Estimator에서 ROC AUC 수치가 어떻게 향상되는 지 확인

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 향상시키기 위해 n_estimators를 100으로 감소
xgb_clf = XGBClassifier(n_estimators=100, random_state=156)

params = {
    'max_depth': [5, 7],
    'min_child_weight': [1, 3],
    'colsample_bytree': [0.25, 0.75]
}

# 하이퍼 파라미터 테스트의 수행속도를 향상시키기 위해 cv를 지정하지 않음
gridcv = GridSearchCV(xgb_clf, param_grid=params)
gridcv.fit(X_train, y_train,
           early_stopping_rounds=30,
           eval_metric='auc',
           eval_set=[(X_train, y_train), (X_test, y_test)])

print('GridSearchCV 최적 파라미터 : ', gridcv.best_params_)

xgb_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:, 1], average='macro')
print('ROC AUC : {0:.4f}'.format(xgb_roc_score))
```

$\rightarrow$ 하이퍼 파라미터 튜닝 전 ROC AUC 0.8419에서 0.8447로 개선됨

<br>

최적화 하이퍼 파라미터를 기반으로 다른 하이퍼 파라미터를 변격ㅇ 또는 추가해 다시 최적화를 진행  
위에서 얻은 하이퍼 파라미터들에 더해 아래의 하이퍼 파라미터 추가  
$\rightarrow$ `n_estimators=1000`으로 증가  
$\rightarrow$ `learning_rate=0.02`로 감소(`learning_rate` 기본값은 0.3)  
$\rightarrow$ `reg_alpha=0.03` 추가

```python
xgb_clf = XGBClassifier(n_estimators=1000, random_state=156,
                        learning_rate=0.02,
                        max_depth=5,
                        min_child_weight=1,
                        colsample_bytree=0.25,
                        reg_alpha=0.03)

# 성능 평가 지표를 auc로, 조기 중단 파라미터 값은 200으로 설정하고 학습 수행
xgb_clf.fit(X_train, y_train, early_stopping_rounds=200,
            eval_metric='auc',
            eval_set=[(X_train, y_train), (X_test, y_test)])

xgb_roc_score = roc_auc_score(y_test, xgb_clf.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC : {0:.4f}'.format(xgb_roc_score))
sn.noti('ROC AUC : {0:.4f}'.format(xgb_roc_score))
```

$\rightarrow$ 이전 ROC AUC 0.8447에서 0.8454로 개선됨

<br>

## 3.3 피처 중요도 시각화

튜닝된 모델에서 각 피처의 중요도를 피처 중요도 그래프로 표시  
`xgboost` 모듈의 `plot_importance()` 메서드 이용

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(10, 8))
plot_importance(xgb_clf, ax=ax, max_num_features=20, height=0.4)
```

$\rightarrow$ XGBoost의 예측 성능을 좌우하는 가장 중요한 피처는 var38, var15 순이다.

<br>

# 4. LightGBM 모델 학습과 하이퍼 파라미터 튜닝

## 4.1 모델 학습

XGBoost 예제와 동일하게 하이퍼 파라미터 설정  
$\rightarrow$ `n_estimators=500`  
$\rightarrow$ `early_stopping_rounds=100`  
$\rightarrow$ `eval_set=[(X_test, y_test)]` (테스트 데이터 세트)  
$\rightarrow$ `eval_metric='auc'` 

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score

lgbm_clf = LGBMClassifier(n_estimators=500, random_state=156)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train,
             early_stopping_rounds=500,
             eval_metric='auc',
             eval_set=evals,
             verbose=True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC : {0:.4f}'.format(lgbm_roc_score))
sn.noti('LightGBM AUC : {0:.4f}'.format(lgbm_roc_score))
```

$\rightarrow$ ROC AUC가 약 0.8396을 나타냄  
(XGBoost보다 LightGBM의 학습 속도가 더 빠르다는 것을 확인 가능)

<br>

## 4.2 하이퍼 파라미터 튜닝

튜닝 대상  
$\rightarrow$ `num_leaves`  
$\rightarrow$ `max_depth`  
$\rightarrow$ `min_child_samples`  
$\rightarrow$ `subsample`

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼 파라미터 테스트의 수행 속도를 증가시키기 위해 n_estimators를 200으로 감소
lgbm_clf = LGBMClassifier(n_estimators=200, random_state=156)

params = {
    'num_leaves': [32, 64],
    'max_depth': [128, 160],
    'min_child_samples': [60, 100],
    'subsample': [0.8, 1]
}

# 하이퍼 파라미터 테스트의 수행속도를 향상시키기 위해 cv를 지정하지 않음
gridcv = GridSearchCV(lgbm_clf, param_grid=params)
gridcv.fit(X_train, y_train,
           early_stopping_rounds=30,
           eval_metric='auc',
           eval_set=[(X_train, y_train), (X_test, y_test)])
sn.noti('LightGBM Learning Finished!')

print('GridSearchCV 최적 파라미터 : ', gridcv.best_params_)
lgbm_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC : {0:.4f}'.format(lgbm_roc_score))
sn.noti('LightGBM AUC : {0:.4f}'.format(lgbm_roc_score))
```

$\rightarrow$ ROC AUC가 0.8396에서 0.8442로 개선됨

<br>

해당 하이퍼 파라미터를 LightGBM에 적용하고 다시 학습해 ROC-AUC 측정 결과를 도출

```python
lgbm_clf = LGBMClassifier(n_estimators=1000,
                          num_leaves=32,
                          subsample=0.8,
                          min_child_samples=100,
                          max_depth=128)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train,
             early_stopping_rounds=100,
             eval_metric='auc',
             eval_set=evals,
             verbose=True)
sn.noti('LightGBM Learning Finsihed!')

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:, 1], average='macro')
print('ROC AUC : {0:.4f}'.format(lgbm_roc_score))
sn.noti('LightGBM AUC : {0:.4f}'.format(lgbm_roc_score))
```
